# Title and description of the project

Iron Man Blaster. <br><br>
Here is our initial project idea we want to pursue: We will be designing a glove to simulate an early model of an Iron Man Blaster. This will be a glove that contains LEDs and Buzzers on the palm of your hand, which will go off at the press of a button to simulate a shot. Holding the button will simulate firing continuous shots. When shots fire, a Ultrasonic Distance Sensor attached to the palm of the hand will take in data about the object being fired at. Depending on the distance of the object being fired at, an algorithm we make will calculate the accuracy of the laser throughout the shot. We are assuming the shots will hit a random spot within a set radius, and as the distance between the user and the target increases, that radius will also increase. This simulates losing accuracy as distance increases.

# How the project requirements will be met
Original Project:<br>
Sensors: To meet the hardware requirements we will be using multiple pieces of equipment such as LEDs, Buzzers, Buttons, and Ultrasonic Distance Sensor.<br>

Data Processing: We would be taking in data from the Ultrasonic Distance Sensor and putting that data into numpy arrays. We would be performing calculations on the data collected by the distance sensor to convert this into an accuracy based off our algorithm designs. These arrays would be converted into matplotlib graphs to easily view and analyze how the accuracy changes over the time of a shot. The numpy arrays would be stored in files to save the data. <br>

We will be continously developing Notebooks with our work that will be put together at the end for a cohessive report. This will let us maintain good notes we can reference for our presentation and video.
